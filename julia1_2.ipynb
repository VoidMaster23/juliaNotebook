```
using PyPlot
#setting the trange for t
t1 = -0.5; t2 = 0.5; delT = 0.01;
t = t1: delT : t2

#fill the array with function values
rect = (abs.(t).<= 0.5).*1.0
omega = 2*pi*6
x = 10 *rect.*cos.(omega*t)

#calculate instantaneous power
ipwr = (abs.(x)).^2  #rember the . takes each value and performs the operation

#integrate
e = sum(ipwr)*delT

#plot the ting
figure(); plot(t,x); figure(); plot(t,ipwr)
println(e)
```
